```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df['encoded_target'] = data.target
```

```python
df
```

```python
df.iloc[0,0] = None
df.iloc[0,1] = None
df.iloc[0,2] = None
df.iloc[1,0] = None
df.iloc[4,3] = None
df.iloc[3,3] = -999999999999999999999999
df.iloc[149,2] = "batatinhafrita123"
```

# **Alternativa seria fazer substituição**


1.   **pela média da coluna com mean()** - Este seria o meio mais comum de tratamento mas implica na interferência nos dados, tome cuidado se os seus dados tem escala negativa e positiva
2.   **por zero com** - substituir por zero pode acrescer a média podendo prejudicar sua distribuição.
3.   **por valor impossível** - cria outlayers mas identifica os dados com problema.


# **substitui pela média da coluna**
#### df_media = df.fillna(df.mean())

# **substitui por zero**
#### df_zero = df.fillna(0)

# **substitui por valor impossível**
#### df_impossiveis = df.fillna(-999)



# **Vamos continuar a partir do df_zero**

```python
df_zero = df.fillna(0) # substitui por zero
df = df_zero
df
```

# **Dados corrompidos**


*  Verificamos por df.info() que a coluna   **petal length (cm)**  é    **object** .

Desta forma simulamos um registro corrompido para exemplificar como encontramos estes registros corrompidos.




```python
df.info()
```

# **Identificando os dados corrompidos**

```python
df['petal length (cm)'].unique()
```

# **Dropando registros corrompidos**

```python
df = df[df['petal length (cm)'] != 'batatinhafrita123'].reset_index(drop=True)
df
```

```python
df['petal length (cm)'].unique()
```

# **Verificando pontos fora da curva**

Abaixo podemos ver -1.0e+24 foi o número -999999999999999999999999 que inserimos no início propositadamente.

```python
df['petal width (cm)'].unique()
```

# **Removendo pontos fora da curva**

```python
df = df[df['petal width (cm)']!= -1.0e+24]
df['petal width (cm)'].unique()
```

# **adicionando a coluna artificial**

```python
import numpy as np

df['artificial'] = np.random.uniform(100,1000,df.shape[0]).round(2)
```

# **Regularização de Dados**



*   Vamos fazer a regularização MinMaxScaler para que todos os valores mantenham suas proporções entre 0 e 1 o que facilita o treinamento de máquina.




```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

print(scaler.fit(df))

novo_df = scaler.transform(df)
```

```python
novo_df = pd.DataFrame(novo_df, columns = df.columns).round(2)
novo_df
```
